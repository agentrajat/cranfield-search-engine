## Importing required libraries

## Initializing IR System

```python
tf_types = ["Binary","Raw frequency","Log normalization","Double normalization 0.5","Double normalization K = 0.4"]
idf_types = ["Unary","Inverse frequency","Inverse frequency with log","Inverse frequency smooth","Inverse frequency max","Probabilistic inverse frequency"]
double_normalization_k = 0.4

table = pd.DataFrame(columns=idf_types, index=tf_types)

for tf_variation in range(1, 6):
    for idf_variation in range(1, 7):
        ir_system = IRSystem(initialize_bm25=False,
                     initialize_qldps=False,
                     vsm_args={
                        "tf_variation": tf_variation,
                        "idf_variation": idf_variation,
                        "double_normalization_k": double_normalization_k
                     })
        evaluator = ir_system.evaluate_model(ir_system.get_model("vsm"))
        results = ir_system.get_metrics(evaluator)
        table.iloc[tf_variation-1, idf_variation-1] = results["map"]
        print(f"TF: {tf_types[tf_variation-1]}, IDF: {idf_types[idf_variation-1]} - MAP: {results['map']}")
```
