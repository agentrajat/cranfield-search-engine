## Initializing IR System

```python
import numpy as np
normalization = [str(round(k, 1)) for k in np.arange(0.1, 1.0, 0.1)]
idf_types = ["Unary","Inverse frequency","Inverse frequency with log","Inverse frequency smooth","Inverse frequency max","Probabilistic inverse frequency"]

table2 = pd.DataFrame(columns=idf_types, index=normalization)

for index, double_normalization_k in enumerate(normalization):
    for idf_variation in range(1, 7):
        ir_system = IRSystem(initialize_bm25=False,
                     initialize_qldps=False,
                     vsm_args={
                        "tf_variation": 5,
                        "idf_variation": idf_variation,
                        "double_normalization_k": float(double_normalization_k),
                     })
        evaluator = ir_system.evaluate_model(ir_system.get_model("vsm"))
        results = ir_system.get_metrics(evaluator)
        table2.iloc[index, idf_variation-1] = results["map"]
        print(f"Normalization: {float(double_normalization_k)}, IDF: {idf_types[idf_variation-1]} - MAP: {results['map']}")
```
