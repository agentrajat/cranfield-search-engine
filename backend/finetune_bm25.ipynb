The purpose of this notebook is to finetune k and b parameters for BM25 model

## Importing required libraries

```python
from ir_system import IRSystem
from IPython.display import Markdown, display
import pandas as pd
import numpy as np
```

## Initializing IR System

```python
k_range = np.arange(1.2, 2.1, 0.1)
b_range = np.arange(0.5, 0.85, 0.05)

table = pd.DataFrame(columns=k_range, index=b_range)

for k in k_range:
    for b in b_range:
        ir_system = IRSystem(
            initialize_vsm=False,
            initialize_qldps=False,
            bm25_args={
                "k": float(round(k, 1)),
                "b": float(round(b, 2))
            }
        )
        evaluator = ir_system.evaluate_model(ir_system.get_model("bm25"))
        results = ir_system.get_metrics(evaluator)
        table.loc[b, k] = results["map"]
        print(f"k: {k}, b: {b} - MAP: {results['map']}")
```
